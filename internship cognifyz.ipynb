Task 8: Investment Duration

Objective: Calculate the average investment
duration.

```python
def duration_midpoint(Duration):
  if Duration == 'Less than 1 year':
    return 0.5
  if Duration == '1-3 years':
    return 2
  if Duration == '3-5 years':
    return 4
  if Duration == 'More than 5 years':
    return 6
  return None

data["Duration_midpoint"]= data["Duration"].apply(duration_midpoint)
```

```python
df_statistic=statistics.mean(data['Duration_midpoint'])
print(f'The average investment duration is {df_statistic:.2f} years.')

# df_statistic=statistics.median(data['Duration_midpoint'])
# print(f'The median of investment duration is {df_statistic:.2f} years.')

# df_statistic=statistics.mode(data['Duration_midpoint'])
# print(f'The mode of investment duration is {df_statistic:.2f} years.')

# df_statistic=statistics.stdev(data['Duration_midpoint'])
# print(f'The standard deviation from an sample of data of investment duration is {df_statistic:.2f} years.')

# df_statistic=statistics.variance(data['Duration_midpoint'])
# print(f'The variance of investment duration is {df_statistic:.2f} years.')
```

Task 10: Correlation Analysis

```python
# Mapping Expect to numeric values

mapping={'10%-20%': '15','20%-30%': '25','30%-40%': '35'}


data['Expect_return']=data['Expect'].map(mapping).astype(float)
data['Expect_return'].head(10)
```

```python
# Calculate correlation matrix
corr=data[["Duration_midpoint",'age','Expect_return','Mutual_Funds','Equity_Market','Debentures','Government_Bonds','Fixed_Deposits','PPF','Gold']].corr()

# Display the correlation matrix
corr
```

The correlation matrix shows the following relationships between the selected columns:

Age and Investment Duration: 0.052

Age and Expected Returns: 0.026

Investment Duration and Expected Returns: 0.179

These correlations are relatively low, indicating weak linear relationships between these variables.

```python
plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
sns.scatterplot(x=data['Duration'],y=data['Expect'],data=corr,hue=data['gender'],edgecolor="black",linewidth=1,)
plt.title('Duration v/s Expected Returns')

plt.subplot(2,2,2)
sns.scatterplot(x=data['Duration'],y=data['age'],data=corr,hue=data['gender'],edgecolor="black",linewidth=1,)
plt.title('Age v/s Duration')

plt.subplot(2,2,3)
sns.scatterplot(y=data['age'],x=data['Expect'],data=corr,hue=data['gender'],edgecolor="black",linewidth=1,)
plt.title('Age v/s Expected Returns')

plt.show()
```

```python
# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
plt.title('Correlation Heatmap')

sns.heatmap(data=corr,annot=True,cmap="viridis",linewidths=0.9,linecolor='grey',cbar=True,fmt="0.2f",square=False,)
plt.show()
```
